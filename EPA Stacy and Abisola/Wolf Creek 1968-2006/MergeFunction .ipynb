```python
#glob.glob(*.csv) finds all files saved as csv files in the folder where this notebook is saved and prints them out
mergeFiles = glob.glob('*.csv')
mergeFiles
```

```python
#Remember: mergeFiles is just all the csv files in the folder where this is saved
def mergeFile(WolfCreek1968, mergeFiles):
    #for each csv file in mergeFiles
    for file in mergeFiles:
        #read each as a csv file, and read only the first(0), second(1), third(2), and fifth(4) columns
        file = pd.read_csv(file, sep="|", usecols=[0,1,2,4], header=None)
        #rename the 1st column to Tag, 2nd column to Date, 3rd to Parameter Code, and 4th to Values
        file.columns=['Tag','Date','Parameter Code','Values']
        #add leading zeroes to the Parameter Code column since they'll most likely be double digits
        file['Parameter Code'] = file['Parameter Code'].apply(lambda x: '{:0>5}'.format(x))
        #concat basically means 'join together', so join WolfCreek1968 and each csv file vertically (axis = 0)
        fileConcat = pd.concat([WolfCreek1968, file], axis=0)
        #now WolfCreek1968 = WolfCreek1968 + first csv file in the first loop
        #in the next loop, it would be WolfCreek1968 + first csv file = WolfCreek1968 + 1st csv file + 2nd csv file in mergeFiles
        WolfCreek1968 = fileConcat
    #show me WolfCreek1968 + 1st csv + 2nd csv + 3rd csv +..... last csv
    return fileConcat

        
```

```python
AlmostMasterFile = mergeFile(WolfCreek1968, mergeFiles)
#converting the date column to datetime format or there'll be trouble w/ sorting the final dataframe
AlmostMasterFile['Date']=pd.to_datetime(AlmostMasterFile['Date'])
#AlmostMasterFile['Date']
AlmostMasterFile.sort('Date', ascending=[True])

```
