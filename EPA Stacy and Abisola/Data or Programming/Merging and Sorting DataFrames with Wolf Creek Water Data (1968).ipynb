```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import matplotlib
matplotlib.style.use('ggplot')
import csv
```

```python
WolfCreek1968 = pd.read_csv('Wolf Creek Water Data (1968).csv', sep='|')
#renames the columns. Note: MUST go in order 
WolfCreek1968.columns = ['Tag','Date','Parameter Code','Values']
#adds the zeroes in front of the numbers for col 'Parameter Codes' so it merges w/ Parameter_Guide
WolfCreek1968['Parameter Code'] = WolfCreek1968['Parameter Code'].apply(lambda x: '{:0>5}'.format(x))
WolfCreek1968
#WolfCreek1968.to_csv('WolfCreek1968.csv') <---- turns WolfCreek1968 Dataframe back into a csv file 
#WolfCreek1968.to_csv('WolfCreek1968.csv') 


```

```python
#Sorts the columns by Parameter codes first and then by Dates
#Created new variable b/c I wanted to keep original data table (much easier than erasing .sort() all over again)
WolfCreek1968_1= WolfCreek1968.sort(['Parameter Code','Date'], ascending = [True, True])
WolfCreek1968_1
```

```python
#adding the Parameter_Guide in. MUST be done before merging or pandas won't know what Parameter_Guide is 
Parameter_Guide = pd.read_csv("TCEQ Parameter Name Guide.csv", sep='|')
Parameter_Guide
```

```python
#New variable in case I want to use WolfCreek1968_1 again 
#Merging Parameter_Guide w/ WolfCreek1968_1 
WolfCreek1968_Merged = WolfCreek1968_1.merge(Parameter_Guide, on='Parameter Code')
WolfCreek1968_Merged
```

```python
#This time, sorting by Date column first and then by Parameter Code
WolfCreek1968_Merged1 =  WolfCreek1968_Merged.sort(['Date','Parameter Code'], ascending = [True, True])
WolfCreek1968_Merged1
```
